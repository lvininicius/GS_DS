# Carregando o conjunto de dados de emiss√µes de carbono e temperatura globalüìà 

FONTE - *Data on CO2 and Greenhouse Gas Emissions by Our World in Data: https://github.com/owid/co2-data*


FONTE - *Temperature change: https://www.kaggle.com/datasets/sevgisarac/temperature-change*

```python
df = pd.read_csv('owid-co2-data.csv')
df_temperatura_global = pd.read_csv('FAOSTAT_data_en_11-1-2024.csv')
```

## An√°lise Explorat√≥ria dos Dados

Descri√ß√£o do Processo de Limpeza:

Remover dados ausentes e inconsistentes.
Filtrar registros irrelevantes e corrigir valores extremos.
Verificar a consist√™ncia das vari√°veis e remover duplicatas.
Realizar transforma√ß√µes de dados necess√°rias, como normaliza√ß√£o e padroniza√ß√£o.


```python
df = df.dropna()
df = df.drop_duplicates()

threshold = len(df) * 0.5
df_cleaned = df.dropna(thresh=threshold, axis=1)

df_cleaned.fillna(df_cleaned.select_dtypes(include=['float64', 'int64']).mean(), inplace=True)
```

* Calculando os 5 primeiros paises que mais emetiram carbono entre 1990 (Inicio da producao de micro computadores e computadores) ate o ano de 2020


```python
# Definir a lista de maiores emissores (top_emitters)
top_emitters = ['China', 'United States', 'India', 'Russia', 'Japan']

# Certificar-se de que o Brasil est√° inclu√≠do na lista
if 'Brazil' not in top_emitters:
    top_emitters.append('Brazil')

# Filtrar o dataset para incluir apenas os maiores emissores
df_filtered = df[['year', 'country', 'co2']].dropna()
df_top_emitters = df_filtered[df_filtered['country'].isin(top_emitters)]

# Calcular as emiss√µes globais anuais
annual_global_emissions = df_filtered.groupby('year')['co2'].sum()

# Criar uma tabela piv√¥ com emiss√µes dos maiores emissores por ano
df_pivot = df_top_emitters.pivot_table(index='year', columns='country', values='co2', aggfunc='sum')

# Calcular a porcentagem de emiss√µes globais para cada pa√≠s
df_percentage = (df_pivot.div(annual_global_emissions, axis=0)) * 100

# Exibir as 5 primeiras linhas da tabela resultante
print(df_percentage.head())
```

# Plotando Graficos

```python
# Criar o gr√°fico para visualiza√ß√£o das porcentagens
df_percentage.plot(figsize=(14, 8))
plt.title("Porcentagem das Emiss√µes Globais de CO‚ÇÇ pelos Maiores Emissores", fontsize=16)
plt.xlabel("Ano", fontsize=14)
plt.ylabel("Porcentagem das Emiss√µes Globais (%)", fontsize=14)
plt.legend(title="Pa√≠ses", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Avaliando o modelo de classifica√ß√£o


```python

# Combinar os dois datasets com base no ano e no pa√≠s
dados_combinados = pd.merge(
    df,
    df_temperatura_global,
    left_on=['year', 'country'],
    right_on=['Year', 'Area'],
    how='inner'
)

# Renomear as colunas relevantes
dados_combinados.rename(columns={'Value': 'temperature_change', 'co2': 'carbon_emissions'}, inplace=True)

# Converter emiss√µes para milh√µes de toneladas, se necess√°rio
if dados_combinados['carbon_emissions'].max() > 1e6:
    dados_combinados['carbon_emissions'] = dados_combinados['carbon_emissions'] / 1e6

# Remover valores ausentes e aplicar filtro para emiss√µes significativas (>= 10 Mt)
dados_combinados = dados_combinados.dropna(subset=['carbon_emissions', 'temperature_change'])
dados_combinados = dados_combinados[dados_combinados['carbon_emissions'] >= 10]

# Verificar se h√° dados suficientes para an√°lise
if len(dados_combinados) < 2:
    print("Dados insuficientes para calcular a correla√ß√£o ou ajustar o modelo.")
else:
        # Calcular o coeficiente de correla√ß√£o
        r, p_value = pearsonr(dados_combinados['carbon_emissions'], dados_combinados['temperature_change'])
    
        # Ajustar o modelo de regress√£o linear
        X = dados_combinados['carbon_emissions'].values.reshape(-1, 1)
        y = dados_combinados['temperature_change'].values
        modelo = LinearRegression()
        modelo.fit(X, y)
        linha_regressao = modelo.predict(X)
    
        # Ajustar os limites do gr√°fico
        carbon_min, carbon_max = dados_combinados['carbon_emissions'].min(), dados_combinados['carbon_emissions'].max()
        temp_min, temp_max = dados_combinados['temperature_change'].min(), dados_combinados['temperature_change'].max()
        x_min, x_max = 2000, carbon_max + 500  # Ignorar valores abaixo de 2000 Mt
        y_min, y_max = temp_min - 1, temp_max + 1
    
        # Plotar o gr√°fico
        plt.figure(figsize=(12, 6))
        sns.scatterplot(x=dados_combinados['carbon_emissions'], y=dados_combinados['temperature_change'], alpha=0.6, label='Dados Observados')
        plt.plot(dados_combinados['carbon_emissions'], linha_regressao, color='red', label='Linha de Regress√£o')
    
        plt.xlim(x_min, x_max)
        plt.ylim(y_min, y_max)
        plt.title(f"Rela√ß√£o entre Emiss√µes de Carbono e Mudan√ßa de Temperatura\nCoeficiente de Correla√ß√£o (r): {r:.2f}", fontsize=16)
        plt.xlabel("Emiss√µes de CO‚ÇÇ (Mt)", fontsize=14)
        plt.ylabel("Mudan√ßa de Temperatura (¬∞C)", fontsize=14)
        plt.legend(fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()
```

```python
# Definir um limite aceit√°vel de erro (por exemplo, 0.5¬∞C)
limite_erro = 0.5

# Calcular o n√∫mero de previs√µes dentro do limite aceit√°vel
acertos = sum(abs(linha_regressao - y) <= limite_erro)

# Calcular a acur√°cia
acuracia = acertos / len(y) * 100

# Exibir a acur√°ciap
print(f"Acur√°cia do Modelo: {acuracia:.2f}%")
```
