# Carregando o conjunto de dados de emiss√µes de carbono


```python
df = pd.DataFrame({
    'Country': ['Brazil', 'USA', 'Global', 'OECD', 'American Continent'],
    'Year': [2020, 2020, 2020, 2020, 2020],
    'CO2 Emissions (Mt)': [422, 5000, 30000, 12000, 8000],
    'Per Capita Emissions': [2.5, 15.5, 4.8, 9.2, 6.0]
})
```

## An√°lise Explorat√≥ria dos Dados


```python
df.describe()

plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='CO2 Emissions (Mt)', data=df)
plt.title('Emiss√µes de CO2 por Regi√£o (Mt)')
plt.xlabel('Regi√£o')
plt.ylabel('Emiss√µes de CO2 (Mt)')
plt.show()
```
